## Step 1
**First look for projects that may already be doing similar work**

Discovering repositories is not always easy (perhaps AI can help) and there are risks that any projects you find may be old and in disprepair or may be filled with creeps.  However, you may find a fantastic new community that aligns with your passion or you may find code or data that already does much of the work you were planning to do.

> Perhaps researchers may find projects by searching various databases such as Academic Search Complete.

## Step 3

**Create your first repository**

From your Github profile, go to the Repositories tab and hit the green "New" button on the right side near the top, which will open this page.

You may have to authorize Github for VSCode
